```python
df = pd.read_csv("/content/clustering-coefficient.txt", sep=' ')
df
```

```python
print("Number of rows: ", len(df.index)) 
df_1 = df.drop(index = df.index[-2:])
df_2 =df_1.drop(columns=['Node', 'C', '='])
df_2
```

```python
Sorted_CCV = df_2.groupby(["CCV"]).count().sort_values(by="CCV", ascending=False)
Sorted_CCV
```

```python
fig = plt.figure(figsize=(10,6))
plt.title("Clustering Coefficient Value")
plt.scatter(Sorted_CCV, Sorted_CCV, s=30, c='purple', alpha=0.2,)
plt.xlabel('Node Counts')
plt.ylabel('CCV')
plt.show()

```

```python
Removed_CCV = Sorted_CCV.iloc[1:]
Removed_CCV
```

```python
fig = plt.figure(figsize=(10,6))
plt.title("Clustering Coefficient Value")
plt.scatter(Removed_CCV, Removed_CCV, s=30, c='purple', alpha=0.2,)
plt.xlabel('Node Counts')
plt.ylabel('CCV')
plt.show()
```
